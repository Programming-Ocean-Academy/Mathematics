## I. Finite Differences & Numerical Derivative Basics (1–15)

1. Compute forward difference $$\Delta f(x)=f(x+h)-f(x).$$  
2. Compute backward difference $$\nabla f(x)=f(x)-f(x-h).$$  
3. Compute central difference $$\delta f(x)=f(x+\tfrac{h}{2})-f(x-\tfrac{h}{2}).$$  
4. For $$f(x)=x^2,\,h=0.1,$$ compute forward difference at $$x=2.$$  
5. For $$f(x)=\sin x,\,h=0.01,$$ estimate $$f'(0).$$  
6. Approximate $$f'(x)\approx\frac{f(x+h)-f(x-h)}{2h}.$$  
7. Estimate $$f''(x)\approx\frac{f(x+h)-2f(x)+f(x-h)}{h^2}.$$  
8. Given $$f(1)=2, f(1.1)=2.22, f(0.9)=1.82,$$ estimate $$f'(1).$$  
9. Using data points, find $$f'(2)$$ by central difference.  
10. Compute forward-difference table for $$f(x)=x^3$$ with $$h=1.$$  
11. Construct backward difference table for $$f(x)=x^2+1.$$  
12. Derive first derivative formula using Taylor expansion.  
13. Derive second derivative central-difference formula.  
14. Estimate derivative of $$f(x)=e^x$$ at $$x=0$$ using $$h=0.1.$$  
15. Estimate $$f''(x)$$ for $$f(x)=\ln x$$ at $$x=1,\,h=0.01.$$
