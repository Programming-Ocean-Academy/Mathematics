## 1ï¸âƒ£ Tangent & Normal Lines

If \( f'(a) \) exists, the **tangent line** at \( x = a \) is

$$
y - f(a) = f'(a)(x - a)
$$

and the **normal line** (perpendicular) has slope

$$
m_{\text{normal}} = -\frac{1}{f'(a)}
$$

**Example**

For \( f(x) = x^2 \) at \( x = 1 \):

$$
f'(x) = 2x \Rightarrow f'(1) = 2
$$

Tangent line:
$$
y - 1 = 2(x - 1) \Rightarrow y = 2x - 1
$$

Normal line:
$$
y - 1 = -\tfrac{1}{2}(x - 1) \Rightarrow y = -\tfrac{1}{2}x + \tfrac{3}{2}
$$

## 2ï¸âƒ£ Motion Interpretation

If \( s(t) \) = position:

| Quantity | Formula | Meaning |
|-----------|----------|----------|
| Velocity | \( v(t) = s'(t) \) | rate of change of position |
| Acceleration | \( a(t) = v'(t) = s''(t) \) | rate of change of velocity |
| Speed | \( |v(t)| \) | magnitude of velocity |

**Example**

Given \( s(t) = t^3 - 6t^2 + 9t \):
$$
v(t) = 3t^2 - 12t + 9
$$
$$
a(t) = 6t - 12
$$

## 3ï¸âƒ£ Increasing / Decreasing Behavior

- \( f'(x) > 0 \): **increasing**  
- \( f'(x) < 0 \): **decreasing**

**Critical points** occur where \( f'(x) = 0 \) or \( f'(x) \) undefined.

Test intervals around them to classify behavior.

---

## 4ï¸âƒ£ Local Maxima & Minima

### First Derivative Test
- \( f' \) changes from \( + \to - \): **local max**  
- \( f' \) changes from \( - \to + \): **local min**

### Second Derivative Test
$$
f''(a) > 0 \Rightarrow \text{min}, \quad f''(a) < 0 \Rightarrow \text{max}
$$

---

## 5ï¸âƒ£ Concavity & Inflection Points

- \( f''(x) > 0 \): **concave up** (smile)  
- \( f''(x) < 0 \): **concave down** (frown)  
- **Inflection point**: where \( f'' \) changes sign

**Example**
$$
f(x) = x^3 \Rightarrow f''(x) = 6x
$$
Changes sign at \( x = 0 \Rightarrow \text{inflection at } (0,0)
$$

## 6ï¸âƒ£ Optimization Problems

**Process**
1. Identify the quantity to maximize or minimize.  
2. Express it as \( f(x) \) in one variable.  
3. Compute \( f'(x) \) and solve \( f'(x) = 0 \).  
4. Test endpoints or critical points for optimal value.

**Example**

Find max area of a rectangle with perimeter \( 100 \):

$$
A = xy, \quad 2x + 2y = 100 \Rightarrow y = 50 - x
$$
$$
A = x(50 - x) = 50x - x^2
$$
$$
A'(x) = 50 - 2x = 0 \Rightarrow x = 25
$$
Max area \( = 625 \) at \( x = y = 25
$$

## 7ï¸âƒ£ Related Rates

Differentiate both sides of an equation **with respect to time t**.

**Example**
A circle expands with \( A = \pi r^2 \) and \( \frac{dr}{dt} = 3 \):
$$
\frac{dA}{dt} = 2\pi r \frac{dr}{dt} = 6\pi r
$$
At \( r = 5 \):
$$
\frac{dA}{dt} = 30\pi
$$

## 8ï¸âƒ£ Linear Approximation / Differentials

Near \( x = a \):
$$
f(x) \approx f(a) + f'(a)(x - a)
$$

**Example**
Estimate \( \sqrt{4.1} \):

\( f(x) = \sqrt{x} \), \( a = 4 \), \( f'(a) = \frac{1}{2\sqrt{4}} = \frac{1}{4} \)

$$
\sqrt{4.1} \approx 2 + \frac{1}{4}(0.1) = 2.025
$$

## 9ï¸âƒ£ Marginal Analysis (Economics)

If \( C(x) \) = cost, \( R(x) \) = revenue, \( P(x) = R(x) - C(x) \):

| Quantity | Derivative | Meaning |
|-----------|-------------|----------|
| Marginal cost | \( C'(x) \) | cost of one more unit |
| Marginal revenue | \( R'(x) \) | revenue from one more unit |
| Marginal profit | \( P'(x) \) | rate of profit change |
