# Multivariable Calculus Foundations

---

Calculus now expands into **multiple dimensions**, where functions depend on several variables and describe surfaces, fields, and 3D phenomena.

---

## 1Ô∏è‚É£ Functions of Several Variables

A function of two variables:
$$
z = f(x, y)
$$
assigns each point \((x, y)\) a value \(z\), forming a surface above the xy-plane.

| Idea | Example |
|------|----------|
| Plane | \( z = 3x + 2y + 1 \) |
| Paraboloid | \( z = x^2 + y^2 \) |
| Saddle | \( z = x^2 - y^2 \) |

---

## 2Ô∏è‚É£ Limits & Continuity in 2D

A limit exists if the value is the same along every path:
$$
\lim_{(x,y)\to(a,b)} f(x,y) = L
$$

Continuity means:
$$
\lim_{(x,y)\to(a,b)} f(x,y) = f(a,b)
$$

---

## 3Ô∏è‚É£ Partial Derivatives

Differentiate with respect to one variable while keeping others constant:
$$
f_x(x,y) = \frac{\partial f}{\partial x}, \quad
f_y(x,y) = \frac{\partial f}{\partial y}
$$

**Example**
$$
f(x,y) = x^2y + 3y^2 \Rightarrow f_x = 2xy, \ f_y = x^2 + 6y
$$

---

## 4Ô∏è‚É£ Higher and Mixed Partials

$$
f_{xx} = \frac{\partial^2 f}{\partial x^2}, \quad
f_{yy} = \frac{\partial^2 f}{\partial y^2}, \quad
f_{xy} = \frac{\partial^2 f}{\partial y \partial x}
$$

If \( f \) is smooth, then \( f_{xy} = f_{yx} \) (Clairaut‚Äôs Theorem).

---

## 5Ô∏è‚É£ Tangent Plane & Linear Approximation

At point \((a, b, f(a,b))\):
$$
z - f(a,b) = f_x(a,b)(x - a) + f_y(a,b)(y - b)
$$

This represents the **tangent plane** to \( z = f(x, y) \).

**Example**
$$
f(x,y) = x^2 + y^2 \text{ at } (1,2):
$$
$$
f_x = 2x, \ f_y = 2y \Rightarrow z - 5 = 2(x - 1) + 4(y - 2) \Rightarrow z = 2x + 4y - 5
$$

---

## 6Ô∏è‚É£ Gradient Vector

$$
\boxed{\nabla f = \langle f_x, f_y, f_z \rangle}
$$

*Direction of steepest ascent:* direction of \( \nabla f \).  
*Rate of change:* \( |\nabla f| \).

**Directional Derivative**
$$
D_{\mathbf{u}} f = \nabla f \cdot \mathbf{u} = |\nabla f| \cos\theta
$$

---

## 7Ô∏è‚É£ Critical Points & Optimization

Set partial derivatives equal to zero:
$$
f_x = 0, \quad f_y = 0
$$

Then test with the **Second Derivative Test**:
$$
D = f_{xx} f_{yy} - (f_{xy})^2
$$

| Condition | Type |
|------------|------|
| \( D > 0, f_{xx} > 0 \) | Local Min |
| \( D > 0, f_{xx} < 0 \) | Local Max |
| \( D < 0 \) | Saddle Point |
| \( D = 0 \) | Inconclusive |

**Example**
$$
f(x,y) = x^2 + y^2 \Rightarrow f_x = 2x, f_y = 2y \Rightarrow (0,0)
$$
$$
f_{xx} = f_{yy} = 2 > 0 \Rightarrow \text{Minimum at } (0,0)
$$

---

## 8Ô∏è‚É£ Lagrange Multipliers (Constrained Optimization)

To optimize \( f(x,y) \) under constraint \( g(x,y) = k \):
$$
\nabla f = \lambda \nabla g
$$

**Example**
Maximize \( f = x + y \) with \( x^2 + y^2 = 1 \):
$$
\nabla f = (1,1), \quad \nabla g = (2x,2y)
$$
$$
1 = 2\lambda x, \ 1 = 2\lambda y \Rightarrow x = y = \pm \frac{1}{\sqrt{2}}
$$
Max at \( (\tfrac{1}{\sqrt{2}}, \tfrac{1}{\sqrt{2}}) \), Min at \( (-\tfrac{1}{\sqrt{2}}, -\tfrac{1}{\sqrt{2}}) \).

---

## 9Ô∏è‚É£ Double & Triple Integrals

### (a) Double Integral
$$
\boxed{\iint_R f(x,y)\,dA = \int_a^b \int_c^d f(x,y)\,dy\,dx}
$$
Gives **volume under a surface**.

---

### (b) Triple Integral
$$
\iiint_V f(x,y,z)\,dV
$$
Measures **mass**, **charge**, or **volume** in 3D.

---

### (c) Average Value
$$
f_{\text{avg}} = \frac{1}{A(R)} \iint_R f(x,y)\,dA
$$
