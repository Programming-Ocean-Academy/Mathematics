# Applied Integration (Area, Volume, Work, Average Value, Probability)

---

Integration connects directly to **geometry, physics, economics, and probability**.  
It reveals how accumulated quantities like area, volume, work, and expectation arise from continuous change.

## 1Ô∏è‚É£ Area Between Curves

If \( f(x) \ge g(x) \) on \( [a,b] \):
$$
\boxed{A = \int_a^b [f(x) - g(x)]\,dx}
$$

**Steps**

1. Graph or determine which curve is on top.  
2. Integrate the vertical distance \( f - g \).

**Example**
$$
A = \int_0^2 (x^2 - (x + 1))\,dx
= \left[\frac{x^3}{3} - \frac{x^2}{2} - x\right]_0^2
= \frac{8}{3} - 2 - 2 = \frac{2}{3}.
$$

## 2Ô∏è‚É£ Volume of Solids of Revolution

Rotating a region around an axis forms a solid.

### (a) Disk Method ‚Äî rotation about x-axis
$$
\boxed{V = \pi \int_a^b [f(x)]^2\,dx}
$$

### (b) Washer Method ‚Äî rotation between two curves
$$
\boxed{V = \pi \int_a^b \big([R(x)]^2 - [r(x)]^2\big)\,dx}
$$

**Example**  
Region under \( y = \sqrt{x} \), \( x \in [0,4] \), revolved about x-axis:
$$
V = \pi \int_0^4 x\,dx = 8\pi.
$$

### (c) Shell Method ‚Äî rotation about y-axis
$$
\boxed{V = 2\pi \int_a^b x f(x)\,dx}
$$

## 3Ô∏è‚É£ Work Done by a Variable Force

If \( F(x) \) is a force (N) over displacement \( x \) (m):
$$
\boxed{W = \int_a^b F(x)\,dx}
$$

**Example**  
Stretching a spring (Hooke‚Äôs Law \( F = kx \)):
$$
W = \int_0^x kx\,dx = \frac{1}{2}k x^2.
$$

## 4Ô∏è‚É£ Center of Mass (1D)

For a thin rod with density \( \rho(x) \):
$$
\bar{x} = \frac{\int_a^b x\rho(x)\,dx}{\int_a^b \rho(x)\,dx}
$$

## 5Ô∏è‚É£ Average Value of a Function

Over \( [a,b] \):
$$
\boxed{f_{\text{avg}} = \frac{1}{b-a} \int_a^b f(x)\,dx}
$$

**Example**
$$
f(x) = x^2, \ [0,3] \Rightarrow
f_{\text{avg}} = \frac{1}{3}\int_0^3 x^2\,dx
= \frac{1}{3} \cdot 9 = 3.
$$

---

## 6Ô∏è‚É£ Arc Length of a Curve

For \( y = f(x) \) on \( [a,b] \):
$$
\boxed{L = \int_a^b \sqrt{1 + [f'(x)]^2}\,dx}
$$

**Example**
For \( f(x) = x^{3/2}, \ 0 \le x \le 1 \):
\( f'(x) = \frac{3}{2}\sqrt{x} \)
$$
L = \int_0^1 \sqrt{1 + \frac{9}{4}x}\,dx
$$

---

## 7Ô∏è‚É£ Surface Area of Revolution

About x-axis:
$$
\boxed{S = 2\pi \int_a^b f(x)\sqrt{1 + [f'(x)]^2}\,dx}
$$

---

## 8Ô∏è‚É£ Probability and Expected Value

For continuous random variable \( X \) with pdf \( p(x) \):

| Quantity | Formula |
|-----------|----------|
| Probability on [a,b] | \( P(a \le X \le b) = \int_a^b p(x)\,dx \) |
| Expected Value | \( E[X] = \int_{-\infty}^{\infty} x\,p(x)\,dx \) |

**Example**  
Uniform distribution on [0,1]: \( p(x) = 1 \)
$$
E[X] = \int_0^1 x\,dx = \frac{1}{2}.
$$
