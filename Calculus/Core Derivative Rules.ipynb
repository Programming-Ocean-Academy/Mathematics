# Phase 2 — Core Derivative Rules

---

## 1️⃣ Power Rule

If \( f(x) = x^n \), then  
$$
\boxed{f'(x) = n x^{n-1}}
$$

Works for all real \( n \) — positive, negative, or fractional.

**Examples**
$$
f(x) = x^3 \Rightarrow f'(x) = 3x^2
$$
$$
f(x) = x^{-2} \Rightarrow f'(x) = -2x^{-3}
$$
$$
f(x) = \sqrt{x} = x^{1/2} \Rightarrow f'(x) = \tfrac{1}{2}x^{-1/2} = \frac{1}{2\sqrt{x}}
$$

---

## 2️⃣ Constant Rule & Constant Multiple Rule

$$
\frac{d}{dx}(c) = 0
$$

$$
\frac{d}{dx}[c\,f(x)] = c\,f'(x)
$$

**Examples**
$$
\frac{d}{dx}(7) = 0
$$
$$
\frac{d}{dx}(5x^3) = 5 \cdot 3x^2 = 15x^2
$$

---

## 3️⃣ Sum and Difference Rule

$$
\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x)
$$

**Example**
$$
\frac{d}{dx}(x^3 + 4x^2 - 7x) = 3x^2 + 8x - 7
$$

---

## 4️⃣ Product Rule

When functions multiply:
$$
\boxed{(fg)' = f'g + fg'}
$$

**Example**
$$
f(x) = x^2, \quad g(x) = \sin x
$$
$$
f'(x) = 2x, \quad g'(x) = \cos x
$$
$$
\Rightarrow (x^2 \sin x)' = 2x\sin x + x^2\cos x
$$

---

## 5️⃣ Quotient Rule

For a quotient \( \frac{f(x)}{g(x)} \):
$$
\boxed{\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}}
$$

**Example**
$$
f(x) = x^2, \quad g(x) = x + 1
$$
$$
f'(x) = 2x, \quad g'(x) = 1
$$
$$
\Rightarrow \left(\frac{x^2}{x+1}\right)' = \frac{(2x)(x+1) - x^2(1)}{(x+1)^2}
= \frac{x^2 + 2x}{(x+1)^2}
$$

---

## 6️⃣ Chain Rule

For a composition \( f(g(x)) \):
$$
\boxed{(f \circ g)' = f'(g(x)) \cdot g'(x)}
$$

**Example**
$$
f(x) = (3x^2 + 1)^5
$$
Let \( u = 3x^2 + 1 \).  
Then \( f(x) = u^5 \Rightarrow f'(x) = 5u^4 \cdot \frac{du}{dx} = 5(3x^2 + 1)^4 \cdot 6x = 30x(3x^2 + 1)^4
$$

---

## 7️⃣ Derivatives of Common Functions

| Function | Derivative |
|-----------|-------------|
| \( c \) | \( 0 \) |
| \( x^n \) | \( n x^{n-1} \) |
| \( e^x \) | \( e^x \) |
| \( a^x \) | \( a^x \ln a \) |
| \( \ln x \) | \( \frac{1}{x} \) |
| \( \sin x \) | \( \cos x \) |
| \( \cos x \) | \( -\sin x \) |
| \( \tan x \) | \( \sec^2 x \) |
| \( \sec x \) | \( \sec x \tan x \) |
| \( \csc x \) | \( -\csc x \cot x \) |
| \( \cot x \) | \( -\csc^2 x \) |

---

## 8️⃣ Combined Example

Differentiate:
$$
y = e^{3x}\sin(2x)
$$

**Solution**

Use product + chain rules:
$$
y' = e^{3x}(3\sin(2x) + 2\cos(2x))
$$

## 9️⃣ Summary Sheet

| Rule | Formula | Use When |
|-------|----------|-----------|
| Power | \( n x^{n-1} \) | \( x^n \) |
| Constant | \( 0 \) | constants |
| Constant Multiple | \( c f'(x) \) | constant × function |
| Sum/Diff | \( f' \pm g' \) | addition/subtraction |
| Product | \( f'g + fg' \) | multiplication |
| Quotient | \( \frac{f'g - fg'}{g^2} \) | division |
| Chain | \( f'(g(x)) \cdot g'(x) \) | nested functions |
