##  A. Derivative Rules

---

### 1Ô∏è‚É£ Power Rule
$$
\frac{d}{dx}(x^n) = n x^{n-1}
$$

---

### 2Ô∏è‚É£ Constant Rule
$$
\frac{d}{dx}(c) = 0
$$

---

### 3Ô∏è‚É£ Constant Multiple Rule
$$
\frac{d}{dx}[c \cdot f(x)] = c \cdot f'(x)
$$

---

### 4Ô∏è‚É£ Sum / Difference Rule
$$
\frac{d}{dx}[f(x) \pm g(x)] = f'(x) \pm g'(x)
$$

---

### 5Ô∏è‚É£ Product Rule
$$
(fg)' = f'g + fg'
$$

---

### 6Ô∏è‚É£ Quotient Rule
$$
\left(\frac{f}{g}\right)' = \frac{f'g - fg'}{g^2}
$$

---

### 7Ô∏è‚É£ Chain Rule
$$
\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)
$$

---

### 8Ô∏è‚É£ Exponential Functions
$$
\frac{d}{dx}(e^x) = e^x, \quad
\frac{d}{dx}(a^x) = a^x \ln a
$$

---

### 9Ô∏è‚É£ Logarithmic Functions
$$
\frac{d}{dx}(\ln x) = \frac{1}{x}, \quad
\frac{d}{dx}(\log_a x) = \frac{1}{x \ln a}
$$

---

### üîü Trigonometric Derivatives

| \( \tan x \) | \( \sec^2 x \) |
| \( \cot x \) | \( -\csc^2 x \) |

|-----------|-------------|

| \( \tan^{-1}x \) | \( \frac{1}{1 + x^2} \) |
| \( \cot^{-1}x \) | \( -\frac{1}{1 + x^2} \) |
| \( \sec^{-1}x \) | \( \frac{1}{|x|\sqrt{x^2 - 1}} \) |
| \( \csc^{-1}x \) | \( -\frac{1}{|x|\sqrt{x^2 - 1}} \) |

---

### 1Ô∏è‚É£2Ô∏è‚É£ Implicit Differentiation

If \( y \) is a function of \( x \):

Example:
$$
x^2 + y^2 = 25 \Rightarrow 2x + 2y \frac{dy}{dx} = 0 \Rightarrow \frac{dy}{dx} = -\frac{x}{y}
$$

---

### 1Ô∏è‚É£3Ô∏è‚É£ Higher-Order Derivatives
$$
f''(x) = \frac{d^2y}{dx^2}, \quad f^{(n)}(x) = \frac{d^n y}{dx^n}
$$

---

### 1Ô∏è‚É£4Ô∏è‚É£ Derivatives of Special Combinations
$$
\frac{d}{dx}[f(x)^{g(x)}] = f(x)^{g(x)} \left[g'(x)\ln f(x) + g(x)\frac{f'(x)}{f(x)}\right]
$$

##  B. Integral Rules (Antiderivatives & Definite Integrals)

---

### 1Ô∏è‚É£ Power Rule
$$
\int x^n \,dx = \frac{x^{n+1}}{n+1} + C, \quad n \ne -1
$$

---

### 2Ô∏è‚É£ Constant Rule
$$
\int c\,dx = c x + C
$$

---

### 3Ô∏è‚É£ Constant Multiple Rule
$$
\int c f(x)\,dx = c \int f(x)\,dx
$$

---

### 4Ô∏è‚É£ Sum / Difference Rule
$$
\int [f(x) \pm g(x)]\,dx = \int f(x)\,dx \pm \int g(x)\,dx
$$

---

### 5Ô∏è‚É£ Exponential Integrals
$$
\int e^x\,dx = e^x + C, \quad
\int a^x\,dx = \frac{a^x}{\ln a} + C
$$

---

### 6Ô∏è‚É£ Logarithmic Integral
$$
\int \frac{1}{x}\,dx = \ln|x| + C
$$

---

### 7Ô∏è‚É£ Trigonometric Integrals

| Function | Integral |
|-----------|-----------|
| \( \int \sin x\,dx \) | \( -\cos x + C \) |
| \( \int \cos x\,dx \) | \( \sin x + C \) |
| \( \int \sec^2 x\,dx \) | \( \tan x + C \) |
| \( \int \csc^2 x\,dx \) | \( -\cot x + C \) |
| \( \int \sec x \tan x\,dx \) | \( \sec x + C \) |
| \( \int \csc x \cot x\,dx \) | \( -\csc x + C \) |

---

### 8Ô∏è‚É£ Inverse Trig Integrals

| Function | Integral |
|-----------|-----------|
| \( \int \frac{1}{\sqrt{1 - x^2}}\,dx \) | \( \sin^{-1}x + C \) |
| \( \int \frac{1}{1 + x^2}\,dx \) | \( \tan^{-1}x + C \) |
| \( \int \frac{1}{x\sqrt{x^2 - 1}}\,dx \) | \( \sec^{-1}x + C \) |

---

### 9Ô∏è‚É£ Integration by Substitution
If \( u = g(x) \):
$$
\int f(g(x)) g'(x)\,dx = \int f(u)\,du
$$

---

### üîü Integration by Parts
$$
\int u\,dv = uv - \int v\,du
$$

---

### 1Ô∏è‚É£1Ô∏è‚É£ Partial Fractions
Decompose:
$$
\frac{P(x)}{Q(x)} = \frac{A}{x + a} + \frac{B}{x + b} + \dots
$$
Then integrate term by term.

---

### 1Ô∏è‚É£2Ô∏è‚É£ Definite Integral Properties

| Property | Rule |
|-----------|------|
| Additivity | \( \int_a^b f + \int_b^c f = \int_a^c f \) |
| Reversal | \( \int_b^a f = -\int_a^b f \) |
| Constant multiple | \( \int_a^b c f = c \int_a^b f \) |
| Zero interval | \( \int_a^a f = 0 \) |

---

### 1Ô∏è‚É£3Ô∏è‚É£ Fundamental Theorem of Calculus

1Ô∏è‚É£ Derivative of an integral:
$$
\frac{d}{dx}\left(\int_a^x f(t)\,dt\right) = f(x)
$$

2Ô∏è‚É£ Definite integral evaluation:
$$
\int_a^b f(x)\,dx = F(b) - F(a)
$$

---

### 1Ô∏è‚É£4Ô∏è‚É£ Improper Integrals
Handle infinite bounds or singularities via limits:
$$
\int_a^{\infty} f(x)\,dx = \lim_{b\to\infty}\int_a^b f(x)\,dx
$$

---

### 1Ô∏è‚É£5Ô∏è‚É£ Common Shortcuts

| Type | Formula |
|------|----------|
| \( \int e^{kx}\,dx \) | \( \frac{1}{k}e^{kx} + C \) |

| \( \int \cos(kx)\,dx \) | \( \frac{1}{k}\sin(kx) + C \) |

| \( \int (ax+b)^n\,dx \) | \( \frac{(ax+b)^{n+1}}{a(n+1)} + C \) |
