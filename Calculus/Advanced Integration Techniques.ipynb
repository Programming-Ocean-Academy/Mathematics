# Advanced Integration Techniques

---

Integration becomes both **strategic** and **creative** at this stage.  
This phase teaches how to handle integrals that **don’t fit basic formulas** — by decomposing, substituting, or transforming them cleverly.

---

## 1️⃣ Integration by Parts

### Rule

$$
\boxed{\int u\,dv = uv - \int v\,du}
$$
Derived from the product rule:
$$
(uv)' = u'v + uv'
$$

---

### How to Use

1. Identify parts of the product:  
   Choose \( u \) = easily differentiable part, \( dv \) = easily integrable part.  
2. Compute \( du \) and \( v \).  
3. Apply \( uv - \int v\,du \).

---

### LIATE Rule (priority for choosing \(u\))

**L**ogarithmic → **I**nverse trig → **A**lgebraic → **T**rig → **E**xponential

---

### Example 1

$$
\int x e^x\,dx
$$
Let \( u = x \Rightarrow du = dx \), \( dv = e^x\,dx \Rightarrow v = e^x \).

$$
\Rightarrow xe^x - \int e^x\,dx = e^x(x - 1) + C
$$

✅ **Answer:** \( e^x(x - 1) + C \)

---

### Example 2

$$
\int \ln x\,dx
$$
Let \( u = \ln x \Rightarrow du = \frac{1}{x}dx \), \( dv = dx \Rightarrow v = x \).

$$
\Rightarrow x\ln x - \int x \cdot \frac{1}{x}\,dx = x\ln x - x + C
$$

✅ **Answer:** \( x(\ln x - 1) + C \)

## 2️⃣ Trigonometric Integrals

Integrals involving \( \sin, \cos, \tan, \sec \), etc.

---

### Common Reductions

| Type | Strategy |
|------|-----------|
| \( \sin^m x \cos^n x \) | If one exponent is odd → save one and use \( \sin^2 + \cos^2 = 1 \) |
| \( \tan^m x \sec^n x \) | If sec exponent is even → save \( \sec^2 x \) and use \( \tan^2 + 1 = \sec^2 \) |

---

### Example

$$
\int \sin^3 x \cos x\,dx
$$
Let \( u = \sin x \Rightarrow du = \cos x\,dx \).

$$
\int u^3\,du = \frac{u^4}{4} + C = \frac{\sin^4 x}{4} + C
$$

---

## 3️⃣ Trig Substitution

Used for expressions like \( \sqrt{a^2 - x^2}, \sqrt{a^2 + x^2}, \sqrt{x^2 - a^2} \).

| Expression | Substitution | Identity Used |
|-------------|---------------|----------------|
| \( \sqrt{a^2 - x^2} \) | \( x = a\sin\theta \) | \( 1 - \sin^2 = \cos^2 \) |
| \( \sqrt{a^2 + x^2} \) | \( x = a\tan\theta \) | \( 1 + \tan^2 = \sec^2 \) |
| \( \sqrt{x^2 - a^2} \) | \( x = a\sec\theta \) | \( \sec^2 - 1 = \tan^2 \) |

---

### Example

$$
\int \frac{dx}{\sqrt{1 - x^2}}
$$
Let \( x = \sin\theta \), \( dx = \cos\theta\,d\theta \).

$$
\int \frac{\cos\theta\,d\theta}{\sqrt{1 - \sin^2\theta}} = \int d\theta = \theta + C = \arcsin x + C
$$

✅ **Answer:** \( \arcsin x + C \)

## 4️⃣ Partial Fractions

Used for rational functions where numerator degree < denominator degree.

---

### Steps

1. Factor denominator.  
2. Express as sum of simpler fractions.  
3. Integrate each term separately.

---

### Example

$$
\int \frac{3x + 5}{(x + 1)(x + 2)}\,dx
$$

Write:
$$
\frac{3x + 5}{(x + 1)(x + 2)} = \frac{A}{x + 1} + \frac{B}{x + 2}
$$

Multiply both sides by \( (x + 1)(x + 2) \):
$$
3x + 5 = A(x + 2) + B(x + 1)
$$
Solve: \( A = 2, B = 1 \)

$$
\int\left(\frac{2}{x + 1} + \frac{1}{x + 2}\right)dx = 2\ln|x + 1| + \ln|x + 2| + C
$$

✅ **Answer:** \( 2\ln|x + 1| + \ln|x + 2| + C \)

## 5️⃣ Improper Integrals

When bounds or functions become infinite.

| Type | Example | Check |
|-------|----------|--------|
| Infinite bounds | \( \int_1^{\infty} \frac{1}{x^2}\,dx \) | limit test |
| Vertical asymptote | \( \int_0^1 \frac{1}{\sqrt{x}}\,dx \) | limit near 0 |

---

### Example

$$
\int_1^{\infty} \frac{1}{x^2}\,dx = \lim_{b\to\infty} \left[-\frac{1}{x}\right]_1^b = 1
$$

✅ Convergent integral = 1

## 6️⃣ Numerical Integration

Used when no closed form exists — approximate using areas.

| Method | Formula | Meaning |
|---------|----------|----------|
| Left/Right Riemann | \( \sum f(x_i)\Delta x \) | Rectangular approximation |
| Trapezoidal Rule | \( T_n = \frac{\Delta x}{2}[f(x_0) + 2f(x_1) + \cdots + f(x_n)] \) | Uses trapezoids |
| Simpson’s Rule | \( S_n = \frac{\Delta x}{3}[f(x_0) + 4f(x_1) + 2f(x_2) + \cdots + f(x_n)] \) | Uses parabolic arcs |

---

## 7️⃣ Integration Strategy Summary

| Situation | Recommended Technique |
|-------------|------------------------|
| Simple powers / polynomials | Basic rules |
| Product of functions | Integration by parts |
| Function inside its derivative | u-substitution |
| Rational function | Partial fractions |
| Trig expressions | Trig identities |
| Roots like \( \sqrt{a^2 - x^2} \) | Trig substitution |
| Infinite/undefined region | Improper integral |
| No exact form | Numerical methods |
