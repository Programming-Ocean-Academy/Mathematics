## 2ï¸âƒ£ Gradient (âˆ‡f)

For scalar field \( f(x, y, z) \):
$$
\boxed{\nabla f = \langle f_x, f_y, f_z \rangle}
$$

**Meaning:**  
Direction and rate of **steepest ascent**.  
Perpendicular to level surfaces of \( f \).

**Example**
$$
f = x^2 + y^2 + z^2 \Rightarrow \nabla f = \langle 2x, 2y, 2z \rangle
$$

## 3ï¸âƒ£ Divergence (âˆ‡Â·F)

Measures **net outflow** of a field at a point:
$$
\boxed{\nabla \cdot \mathbf{F} = P_x + Q_y + R_z}
$$

* \( \nabla \cdot \mathbf{F} > 0 \): source (outflow)  
* \( \nabla \cdot \mathbf{F} < 0 \): sink (inflow)  
* \( \nabla \cdot \mathbf{F} = 0 \): incompressible / conserved flow

**Example**
$$
\mathbf{F} = \langle x, y, z \rangle \Rightarrow \nabla \cdot \mathbf{F} = 3
$$

## 4ï¸âƒ£ Curl (âˆ‡Ã—F)

Measures **local rotation** of a field:
$$
\boxed{\nabla \times \mathbf{F} = \langle R_y - Q_z, P_z - R_x, Q_x - P_y \rangle}
$$

If curl = 0 â†’ field is **irrotational**.

**Example**
$$
\mathbf{F} = \langle -y, x, 0 \rangle \Rightarrow \nabla \times \mathbf{F} = \langle 0, 0, 2 \rangle
$$

## 5ï¸âƒ£ Conservative Fields & Potential Functions

A vector field is **conservative** if:
$$
\nabla \times \mathbf{F} = 0
$$
Then there exists a scalar potential \( f \) such that:
$$
\mathbf{F} = \nabla f
$$

The **line integral** between points A and B depends only on endpoints:
$$
\int_C \mathbf{F} \cdot d\mathbf{r} = f(B) - f(A)
$$

## 6ï¸âƒ£ Line Integrals

For curve \( C \) parameterized by \( \mathbf{r}(t) \), \( a \le t \le b \):

**(a) Scalar field (mass along curve):**
$$
\int_C f(x,y,z)\,ds = \int_a^b f(\mathbf{r}(t))\,|\mathbf{r}'(t)|\,dt
$$

**(b) Vector field (work done):**
$$
\boxed{\int_C \mathbf{F} \cdot d\mathbf{r} = \int_a^b \mathbf{F}(\mathbf{r}(t)) \cdot \mathbf{r}'(t)\,dt}
$$

**Example**
$$
\mathbf{F} = \langle y, x \rangle, \quad C: x=t, y=t^2, \ 0 \le t \le 1
$$
Then:
$$
\mathbf{r}'(t) = \langle 1, 2t \rangle, \quad \mathbf{F} \cdot \mathbf{r}' = y(1) + x(2t) = t^2 + 2t^2 = 3t^2
$$
$$
\int_0^1 3t^2\,dt = 1
$$
âœ… Work = 1 unit.
