## 1ï¸âƒ£ Definition

If  
$$
f'(x) = \frac{d}{dx}f(x)
$$
then the **second derivative** is  
$$
f''(x) = \frac{d}{dx}\bigl(f'(x)\bigr) = \frac{d^2y}{dx^2}.
$$

Higher-order derivatives follow similarly:
$$
f^{(3)}(x), \ f^{(4)}(x), \ \dots, \ f^{(n)}(x).
$$

## 2ï¸âƒ£ Meaning

| Order | Symbol | Interpretation |
|:------|:--------|:----------------|
| 1st | \( f'(x) \) | **Rate of change** / slope / velocity |
| 2nd | \( f''(x) \) | **Rate of change of the rate** / curvature / acceleration |
| 3rd+ | \( f^{(n)}(x) \) | Higher-order effects: â€œjerkâ€, curvature change, or dynamics |

## 4ï¸âƒ£ Applications

### (a) Motion Analysis

If \( s(t) \) = position, then:
$$
v(t) = s'(t), \quad a(t) = s''(t), \quad j(t) = s^{(3)}(t)
$$
Here, \( j(t) \) = **jerk**, the rate of change of acceleration.

---

### (b) Concavity & Inflection

- \( f''(x) > 0 \): **concave up** (â€œcupâ€ shape)
- \( f''(x) < 0 \): **concave down** (â€œcapâ€ shape)
- When \( f'' \) changes sign â†’ **inflection point**

---

### (c) Curvature and Speed

For motion along a curve \( y = f(x) \):

$$
\kappa = \frac{|f''(x)|}{(1 + [f'(x)]^2)^{3/2}}
$$

Here \( \kappa \) is **curvature**, a measure of how sharply the curve bends.

---

### (d) Taylor Polynomials

Approximate \( f(x) \) near \( x = a \) using higher-order derivatives:
$$
f(x) \approx f(a)
+ f'(a)(x - a)
+ \frac{f''(a)}{2!}(x - a)^2
+ \frac{f^{(3)}(a)}{3!}(x - a)^3
+ \cdots
$$

**Example**

For \( f(x) = e^x \) at \( a = 0 \):
$$
e^x \approx 1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots
$$
