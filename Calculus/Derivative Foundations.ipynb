## 1ï¸âƒ£ Definition (Formal Limit Definition)

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

This is called the **difference quotient definition of the derivative.**

It expresses how \( f(x) \) changes as the input \( x \) changes by an infinitesimally small amount \( h \).

| Concept | Description |

## 4ï¸âƒ£ Examples (Evaluate by Definition)

### Example 1
$$
f(x) = x^2
$$

$$
\begin{aligned}
f'(x) &= \lim_{h \to 0} \frac{(x + h)^2 - x^2}{h} \\
&= \lim_{h \to 0} \frac{x^2 + 2xh + h^2 - x^2}{h} \\
&= \lim_{h \to 0} (2x + h) \\
&= 2x
\end{aligned}
$$

âœ… \( f'(x) = 2x \)

---

### Example 2
$$
f(x) = 3x + 5
$$

$$
\begin{aligned}
f'(x) &= \lim_{h \to 0} \frac{3(x+h) + 5 - (3x + 5)}{h} \\
&= \lim_{h \to 0} \frac{3h}{h} \\
&= 3
\end{aligned}
$$

âœ… \( f'(x) = 3 \)

### Example 3
$$
f(x) = \sqrt{x}
$$

$$
f'(x) = \lim_{h \to 0} \frac{\sqrt{x + h} - \sqrt{x}}{h}
$$

Multiply numerator and denominator by the conjugate \( \sqrt{x + h} + \sqrt{x} \):

$$
\begin{aligned}
f'(x) &= \lim_{h \to 0}
\frac{(\sqrt{x + h} - \sqrt{x})(\sqrt{x + h} + \sqrt{x})}
{h(\sqrt{x + h} + \sqrt{x})} \\
&= \lim_{h \to 0} \frac{h}{h(\sqrt{x + h} + \sqrt{x})} \\
&= \lim_{h \to 0} \frac{1}{\sqrt{x + h} + \sqrt{x}} \\
&= \frac{1}{2\sqrt{x}}
\end{aligned}
$$

âœ… \( f'(x) = \frac{1}{2\sqrt{x}} \)
