## 1 Ordinary Differential Equation (ODE)

An ODE describes deterministic evolution with respect to a single independent variable (usually time).

### General First-Order ODE
$$
\frac{d x(t)}{d t} = f(t, x(t))
$$

### Autonomous ODE
$$
\frac{d x}{d t} = f(x)
$$

### Linear ODE
$$
\frac{d x}{d t} = a x + b
$$

### Higher-Order ODE
$$
\frac{d^n x(t)}{d t^n} =
f\left(
t,
x,
\frac{d x}{d t},
\ldots,
\frac{d^{n-1} x}{d t^{n-1}}
\right)
$$

### Vector-Valued ODE (Dynamical System)
$$
\frac{d x(t)}{d t} = f(x(t), t),
\quad x(t) \in \mathbb{R}^n
$$

## 2 Partial Differential Equation (PDE)

A PDE describes deterministic evolution involving multiple independent variables (e.g., space and time).

### General PDE
$$
F\left(
x,
t,
u,
\frac{\partial u}{\partial t},
\frac{\partial u}{\partial x},
\frac{\partial^2 u}{\partial x^2},
\ldots
\right) = 0
$$

### Time-Evolution PDE
$$
\frac{\partial u(x,t)}{\partial t} = L u(x,t)
$$

where  
$$
L
$$
is a spatial differential operator.

### Linear PDE (Heat Equation)
$$
\frac{\partial u}{\partial t}
=
\alpha \frac{\partial^2 u}{\partial x^2}
$$

### Nonlinear PDE
$$
\frac{\partial u}{\partial t}
=
f(u, \nabla u, \nabla^2 u, x, t)
$$

---

## 3 Stochastic Differential Equation (SDE)

An SDE models systems with randomness, typically driven by Brownian motion.

### Standard Itô SDE
$$
dX_t
=
\mu(X_t, t)\, dt
+
\sigma(X_t, t)\, dW_t
$$

### Integral Form
$$
X_t
=
X_0
+
\int_0^t \mu(X_s, s)\, ds
+
\int_0^t \sigma(X_s, s)\, dW_s
$$

### Multidimensional SDE
$$
dX_t
=
\mu(X_t, t)\, dt
+
\Sigma(X_t, t)\, dW_t
$$

---

## 4 Stochastic Partial Differential Equation (SPDE)

An SPDE combines spatial dynamics with stochastic forcing.

### General SPDE
$$
du(x,t)
=
L u(x,t)\, dt
+
G u(x,t)\, dW_t
$$

where:

$$
L
$$
: deterministic differential operator  

$$
G
$$
: noise operator

### Additive-Noise SPDE
$$
du(x,t)
=
L u(x,t)\, dt
+
\sigma\, dW_t
$$

### Multiplicative-Noise SPDE
$$
du(x,t)
=
L u(x,t)\, dt
+
\sigma(u(x,t))\, dW_t
$$

### Stochastic Heat Equation
$$
\frac{\partial u}{\partial t}
=
\alpha \frac{\partial^2 u}{\partial x^2}
+
\sigma \dot{W}(x,t)
$$

where  
$$
\dot{W}(x,t)
$$
is space–time white noise.
