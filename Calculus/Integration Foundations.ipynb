## 1Ô∏è‚É£ Definition of the Indefinite Integral

If  
$$
F'(x) = f(x),
$$  
then  
$$
\boxed{\int f(x)\,dx = F(x) + C}
$$  
where \( C \) is the **constant of integration**.

Integration is the **inverse** of differentiation:  
$$
\frac{d}{dx}\left(\int f(x)\,dx\right) = f(x).
$$

## 2Ô∏è‚É£ Geometric Meaning

- **Differentiation ‚Üí slope (rate)**  
- **Integration ‚Üí area (accumulation)**  

$$
\text{Area under } y=f(x) \text{ from } a \text{ to } b = \int_a^b f(x)\,dx
$$

---

## 3Ô∏è‚É£ The Definite Integral

$$
\boxed{\int_a^b f(x)\,dx = \lim_{n\to\infty}\sum_{i=1}^n f(x_i^*)\,\Delta x}
$$

Represents the **net area** between the curve and the x-axis over \([a,b]\):

- Above x-axis ‚Üí positive area  
- Below x-axis ‚Üí negative area  

---

## 4Ô∏è‚É£ The Fundamental Theorem of Calculus

If \( F'(x) = f(x) \), then  
$$
\boxed{\int_a^b f(x)\,dx = F(b) - F(a)}
$$  
This theorem **connects differentiation and integration**.

## 5Ô∏è‚É£ Basic Antiderivative Rules

| Rule | Formula | Example |
|------|----------|----------|
| Power | \( \int x^n\,dx = \tfrac{x^{n+1}}{n+1} + C \) (for \( n \ne -1 \)) | \( \int x^3\,dx = \tfrac{x^4}{4} + C \) |
| Constant | \( \int c\,dx = cx + C \) | \( \int 5\,dx = 5x + C \) |
| Sum/Difference | \( \int(f \pm g)\,dx = \int f\,dx \pm \int g\,dx \) | \( \int(x^2 + 3x)\,dx = \tfrac{x^3}{3} + \tfrac{3x^2}{2} + C \) |
| Exponential | \( \int e^x\,dx = e^x + C,\ \int a^x\,dx = \tfrac{a^x}{\ln a} + C \) | \( \int 2^x\,dx = \tfrac{2^x}{\ln 2} + C \) |
| Trig | \( \int \sin x\,dx = -\cos x + C,\ \int \cos x\,dx = \sin x + C \) | ‚Äî |

## 6Ô∏è‚É£ Integration by Simple Substitution (u-Substitution)

When a function and its derivative appear together:
$$
\boxed{\int f(g(x))\,g'(x)\,dx = \int f(u)\,du}
$$

**Example**
$$
\int 2x(x^2+1)^5\,dx, \quad u = x^2 + 1, \quad du = 2x\,dx
$$
$$
\Rightarrow \int u^5\,du = \tfrac{u^6}{6} + C = \tfrac{(x^2 + 1)^6}{6} + C
$$

## 7Ô∏è‚É£ Properties of Definite Integrals

| Property | Formula |
|-----------|----------|
| Additivity | \( \int_a^b f + \int_b^c f = \int_a^c f \) |
| Reversal | \( \int_b^a f = -\int_a^b f \) |
| Zero Width | \( \int_a^a f = 0 \) |
| Constant Multiple | \( \int_a^b c f = c \int_a^b f \) |

---

## 8Ô∏è‚É£ Example 1 ‚Äî Evaluate a Definite Integral

$$
\int_0^2 (3x^2 + 2x)\,dx = \left[ x^3 + x^2 \right]_0^2 = (8 + 4) - 0 = 12
$$

‚úÖ **Answer:** \( 12 \) (total area under the curve)

---

## 9Ô∏è‚É£ Example 2 ‚Äî Area with Negative Region

$$
\int_{-1}^1 x^3\,dx = \left[ \tfrac{x^4}{4} \right]_{-1}^1 = \tfrac{1}{4} - \tfrac{1}{4} = 0
$$

‚Üí Symmetric odd function ‚áí **net area = 0**
