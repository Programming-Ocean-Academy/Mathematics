# Problem 1 — Power Rule & Linearity (the true “baseline”)

Differentiate:

$$
f(x) = 2x^5 - 3x^3 + 4x - 9
$$

---

### Rules We’ll Use

**Power Rule:**
$$
\frac{d}{dx}x^n = n x^{n-1}
$$

**Constant Multiple Rule:**
$$
\frac{d}{dx}[c \cdot g(x)] = c \cdot g'(x)
$$

**Sum/Difference Rule:**
Differentiate term-by-term.

**Constant Rule:**
$$
\frac{d}{dx}[c] = 0
$$

---

### Step-by-Step

$$
\frac{d}{dx}[2x^5] = 2 \cdot 5x^4 = 10x^4
$$

$$
\frac{d}{dx}[-3x^3] = -3 \cdot 3x^2 = -9x^2
$$

$$
\frac{d}{dx}[4x] = 4 \cdot 1 = 4
$$

$$
\frac{d}{dx}[-9] = 0
$$

---

### Final Answer

$$
f'(x) = 10x^4 - 9x^2 + 4
$$


# Problem 2 — Product Rule

Differentiate:

$$
g(x) = (x^2 + 1)(3x - 2)
$$

---

### Rule Reminder

**Product Rule:**
$$
\frac{d}{dx}[u \cdot v] = u'v + uv'
$$

where  
$$
u = (x^2 + 1), \quad v = (3x - 2)
$$

---

### Step-by-Step

Compute derivatives:
$$
u' = 2x
$$
$$
v' = 3
$$

Apply the product rule:
$$
g'(x) = (2x)(3x - 2) + (x^2 + 1)(3)
$$

Expand each term:
$$
g'(x) = (6x^2 - 4x) + (3x^2 + 3)
$$

Simplify:
$$
g'(x) = 9x^2 - 4x + 3
$$

---

### Final Answer

$$
g'(x) = 9x^2 - 4x + 3
$$


# Problem 3 — Quotient Rule

Differentiate:

$$
h(x) = \frac{x^2 + 1}{x - 3}
$$

---

### Rule Reminder

If  
$$
h(x) = \frac{u}{v},
$$  
then  
$$
h'(x) = \frac{u'v - uv'}{v^2}
$$

where  
$$
u = x^2 + 1, \quad v = x - 3
$$

---

### Step-by-Step

Compute derivatives:
$$
u' = 2x
$$
$$
v' = 1
$$

Substitute into the rule:
$$
h'(x) = \frac{(2x)(x - 3) - (x^2 + 1)(1)}{(x - 3)^2}
$$

Expand the numerator:
$$
h'(x) = \frac{(2x^2 - 6x) - (x^2 + 1)}{(x - 3)^2}
$$

Simplify:
$$
h'(x) = \frac{x^2 - 6x - 1}{(x - 3)^2}
$$

---

### Final Answer

$$
h'(x) = \frac{x^2 - 6x - 1}{(x - 3)^2}
$$


# Problem 4 — Chain Rule

Differentiate:

$$
f(x) = (3x^2 + 2x + 1)^4
$$

---

### Rule Reminder

If  
$$
y = [u(x)]^n,
$$  
then  
$$
\frac{dy}{dx} = n [u(x)]^{n-1} \cdot u'(x)
$$  

This is the **Chain Rule** — multiply the derivative of the outer function by the derivative of the inner function.

---

### Step-by-Step

Let the inner function:  
$$
u = 3x^2 + 2x + 1
$$  
Then:  
$$
u' = 6x + 2
$$

The outer function is \( u^4 \), whose derivative is \( 4u^3 \).

Now apply the rule:  
$$
f'(x) = 4(3x^2 + 2x + 1)^3 \cdot (6x + 2)
$$

Simplify:  
$$
f'(x) = (24x + 8)(3x^2 + 2x + 1)^3
$$

Or equivalently (factored form):  
$$
f'(x) = 4(6x + 2)(3x^2 + 2x + 1)^3
$$

---

### Final Answer

$$
f'(x) = (24x + 8)(3x^2 + 2x + 1)^3
$$


# Problem 5 — Product + Chain Rule

Differentiate:

$$
y = (x^2 + 1)^3 (2x - 5)^4
$$

---

### Setup (Product Rule)

Let  
$$
u = (x^2 + 1)^3, \quad v = (2x - 5)^4
$$

Then  
$$
y' = u'v + uv'
$$

---

### Chain Pieces

Compute \( u' \) using the **Chain Rule**:
$$
u' = 3(x^2 + 1)^2 \cdot (2x) = 6x(x^2 + 1)^2
$$

Compute \( v' \) using the **Chain Rule**:
$$
v' = 4(2x - 5)^3 \cdot 2 = 8(2x - 5)^3
$$

---

### Assemble the Product Rule

Substitute \( u', v, u, v' \):

$$
y' = 6x(x^2 + 1)^2 (2x - 5)^4 + (x^2 + 1)^3 \cdot 8(2x - 5)^3
$$

---

### Factor Common Terms

$$
y' = (x^2 + 1)^2 (2x - 5)^3 [6x(2x - 5) + 8(x^2 + 1)]
$$

Simplify inside the brackets:
$$
y' = (x^2 + 1)^2 (2x - 5)^3 [(12x^2 - 30x) + (8x^2 + 8)]
$$

Combine like terms:
$$
y' = (x^2 + 1)^2 (2x - 5)^3 (20x^2 - 30x + 8)
$$

---

### Final Answer

$$
y' = (x^2 + 1)^2 (2x - 5)^3 (20x^2 - 30x + 8)
$$


# Problem 6 — Implicit Differentiation

Given:

$$
x^2 + y^2 = 25
$$

We must find \( \frac{dy}{dx} \).

---

### Step-by-Step

Differentiate both sides with respect to \( x \), remembering that \( y \) is a function of \( x \):

$$
\frac{d}{dx}(x^2) + \frac{d}{dx}(y^2) = \frac{d}{dx}(25)
$$

Differentiate term by term:

$$
2x + 2y \frac{dy}{dx} = 0
$$

(because \( \frac{d}{dx}(y^2) = 2y \frac{dy}{dx} \) by the chain rule)

---

### Solve for \( \frac{dy}{dx} \)

$$
2y \frac{dy}{dx} = -2x
$$

Divide both sides by \( 2y \):

$$
\frac{dy}{dx} = -\frac{x}{y}
$$

---

### Final Answer

$$
\frac{dy}{dx} = -\frac{x}{y}
$$


# Problem 7 — Logarithmic Differentiation

Differentiate:

$$
y = x^x
$$

---

### Step 1: Take the Natural Log of Both Sides

$$
\ln y = \ln(x^x)
$$

Use the logarithmic identity \( \ln(a^b) = b \ln a \):

$$
\ln y = x \ln x
$$

---

### Step 2: Differentiate Both Sides Implicitly with Respect to \( x \)

$$
\frac{1}{y} \frac{dy}{dx} = 1 \cdot \ln x + x \cdot \frac{1}{x}
$$

Simplify:

$$
\frac{1}{y} \frac{dy}{dx} = \ln x + 1
$$

---

### Step 3: Multiply Both Sides by \( y \)

$$
\frac{dy}{dx} = y (\ln x + 1)
$$

Substitute back \( y = x^x \):

$$
\frac{dy}{dx} = x^x (\ln x + 1)
$$

---

### Final Answer

$$
\frac{dy}{dx} = x^x (\ln x + 1)
$$


# Problem 8 — Derivative of Exponential Function

Differentiate:

$$
f(x) = 5^{2x + 1}
$$

---

### Step 1: Recall the Rule

For any base \( a > 0 \):

$$
\frac{d}{dx}\left(a^{u(x)}\right) = a^{u(x)} \ln(a) \cdot u'(x)
$$

---

### Step 2: Identify the Parts

Here:

$$
a = 5, \quad u(x) = 2x + 1
$$

Then:

$$
u'(x) = 2
$$

---

### Step 3: Apply the Rule

$$
f'(x) = 5^{2x + 1} \ln(5) \cdot 2
$$

Simplify:

$$
f'(x) = 2 \ln(5) \cdot 5^{2x + 1}
$$

---

### Final Answer

$$
f'(x) = 2 \ln(5) \, 5^{2x + 1}
$$

# Problem 19 — Higher-Order Derivatives

Given:

$$
f(x) = x^3 - 6x^2 + 9x + 1
$$

We will find the **first** and **second** derivatives.

---

### Step 1: First Derivative \( f'(x) \)

Differentiate term-by-term:

$$
f'(x) = 3x^2 - 12x + 9
$$

---

### Step 2: Second Derivative \( f''(x) \)

Differentiate \( f'(x) \):

$$
f''(x) = 6x - 12
$$

---

### Final Results

$$
f'(x) = 3x^2 - 12x + 9
$$

$$
f''(x) = 6x - 12
$$

---

### Interpretation

- \( f'(x) \): Represents the **slope** or **rate of change** of \( f(x) \).
- \( f''(x) \): Represents the **curvature** or **concavity** of \( f(x) \).  
  - \( f''(x) > 0 \) ⇒ Concave up  
  - \( f''(x) < 0 \) ⇒ Concave down


# Problem 20 — Critical Points & Local Extrema

Function:
$$
f(x) = x^3 - 3x^2 - 9x + 5
$$

---

## 1) First Derivative & Critical Points
Compute \( f'(x) \):
$$
f'(x) = 3x^2 - 6x - 9 = 3(x^2 - 2x - 3) = 3(x-3)(x+1).
$$

Set \( f'(x) = 0 \) to find critical points:
$$
3(x-3)(x+1)=0 \;\;\Rightarrow\;\; x=-1,\; x=3.
$$

---

## 2) Second Derivative Test
Compute \( f''(x) \):
$$
f''(x) = 6x - 6 = 6(x-1).
$$

At \( x=-1 \):
$$
f''(-1) = 6(-1) - 6 = -12 < 0 \;\Rightarrow\; \text{local maximum}.
$$
Function value:
$$
f(-1) = (-1)^3 - 3(-1)^2 - 9(-1) + 5 = -1 - 3 + 9 + 5 = 10.
$$
Local max at \( (-1,\,10) \).

At \( x=3 \):
$$
f''(3) = 6(3) - 6 = 12 > 0 \;\Rightarrow\; \text{local minimum}.
$$
Function value:
$$
f(3) = 27 - 27 - 27 + 5 = -22.
$$
Local min at \( (3,\,-22) \).

---

## 3) (Optional) Increasing/Decreasing Intervals
Using \( f'(x) = 3(x-3)(x+1) \):
- Increasing on \( (-\infty,-1) \) and \( (3,\infty) \).
- Decreasing on \( (-1,3) \).


# Problem 21 — Concavity & Inflection

Function:
$$
f(x) = x^3 - 3x^2 - 9x + 5
$$

---

## 1) Second Derivative
$$
f''(x) = 6x - 6 = 6(x - 1)
$$

---

## 2) Concavity Intervals
Concave up where \( f''(x) > 0 \):
$$
6(x-1) > 0 \;\;\Longleftrightarrow\;\; x > 1
$$
So, concave up on \( (1,\infty) \).

Concave down where \( f''(x) < 0 \):
$$
6(x-1) < 0 \;\;\Longleftrightarrow\;\; x < 1
$$
So, concave down on \( (-\infty,1) \).

---

## 3) Inflection Point(s)
Solve \( f''(x) = 0 \):
$$
6(x-1)=0 \;\;\Longrightarrow\;\; x=1
$$
Sign changes from negative to positive across \( x=1 \) \( \Rightarrow \) valid inflection point.

Evaluate \( f(1) \):
$$
f(1) = 1 - 3 - 9 + 5 = -6
$$

**Inflection point:** \( (1,-6) \).


# Problem 22 — Optimization (Maximum Product with Fixed Sum)

**Question:** Find two positive numbers whose sum is 20 and whose product is as large as possible.

Let the numbers be \( x \) and \( 20 - x \) with \( 0 < x < 20 \).

**Product function:**
$$
P(x) = x(20 - x) = 20x - x^2
$$

**Differentiate:**
$$
P'(x) = 20 - 2x
$$

**Critical point (set derivative to zero):**
$$
P'(x)=0 \;\Longrightarrow\; 20 - 2x = 0 \;\Longrightarrow\; x = 10
$$
Then the other number is \( 20 - 10 = 10 \).

**Second derivative test:**
$$
P''(x) = -2 < 0 \;\Longrightarrow\; \text{local (and global) maximum on } (0,20).
$$

**Answer:** The two numbers are \( 10 \) and \( 10 \), with maximum product
$$
P_{\max} = 10 \cdot 10 = 100.
$$
