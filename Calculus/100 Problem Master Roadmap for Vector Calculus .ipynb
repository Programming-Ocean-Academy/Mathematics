## I. Vector Algebra & Geometry (1–15)

1. Find magnitude of $$\mathbf{v} = \langle3, -4, 12\rangle.$$
2. Find unit vector in the direction of $$\mathbf{a} = \langle2, -1, 2\rangle.$$
3. Compute $$(3\mathbf{i} - 2\mathbf{j} + \mathbf{k}) + (\mathbf{i} + 4\mathbf{j} - 3\mathbf{k}).$$
4. Find angle between $$\mathbf{a} = \langle1,2,3\rangle$$ and $$\mathbf{b} = \langle3,2,1\rangle.$$
5. Find projection of $$\mathbf{a} = \langle3,4,0\rangle$$ onto $$\mathbf{b} = \langle1,2,2\rangle.$$
6. Compute $$\mathbf{a}\times\mathbf{b}$$ for $$\mathbf{a}=\langle2,1,0\rangle,\mathbf{b}=\langle1,2,3\rangle.$$
7. Find area of parallelogram formed by $$\mathbf{a}$$ and $$\mathbf{b}.$$
8. Compute volume of parallelepiped with $$\mathbf{a}=\langle1,2,3\rangle,\mathbf{b}=\langle2,3,4\rangle,\mathbf{c}=\langle3,4,5\rangle.$$
9. Find scalar triple product $$\mathbf{a}\cdot(\mathbf{b}\times\mathbf{c}).$$
10. Verify $$\mathbf{a}\times\mathbf{b} = -(\mathbf{b}\times\mathbf{a}).$$
11. Find equation of line through (1,2,3) parallel to $$\mathbf{v} = \langle2,-1,4\rangle.$$
12. Find parametric form of line through (0,1,2) and (2,3,4).
13. Find intersection point of line $$(x=1+t,y=2t,z=3-t)$$ with plane $$(x+2y+3z=6).$$
14. Find equation of plane through (1,1,1) with normal $$\langle2,3,4\rangle.$$
15. Find distance from point (1,2,3) to plane $$(2x+y-z=5).$$

## II. Vector Functions & Motion (16–30)

16. $$\mathbf{r}(t)=\langle t, t^2, t^3\rangle;$$ find $$\mathbf{r}'(t).$$  
17. $$\mathbf{r}(t)=\langle\sin t,\cos t, t\rangle;$$ find velocity and acceleration.  
18. Find speed of $$\mathbf{r}(t)=\langle3\cos t, 4\sin t\rangle.$$  
19. Compute tangent and normal vectors for $$\mathbf{r}(t)=\langle t, t^2\rangle.$$  
20. Find unit tangent $$\mathbf{T}(t)$$ for $$\mathbf{r}(t)=\langle e^t, e^{-t}\rangle.$$  
21. Find curvature of $$\mathbf{r}(t)=\langle t, t^2\rangle.$$  
22. Compute acceleration components for $$\mathbf{r}(t)=\langle \cos t, \sin t\rangle.$$  
23. Find normal vector and curvature for $$\mathbf{r}(t)=\langle3t, 4t^2\rangle.$$  
24. Find $$\mathbf{r}'(t)\times\mathbf{r}''(t)$$ for $$\mathbf{r}(t)=\langle t, t^2, t^3\rangle.$$  
25. Find equation of osculating plane at $$t=1$$ for same curve.  
26. Compute torsion for $$\mathbf{r}(t)=\langle t, t^2, t^3\rangle.$$  
27. Compute velocity and acceleration magnitude for projectile $$\mathbf{r}(t)=\langle v_0t, -\tfrac12gt^2\rangle.$$  
28. Determine arc length of $$\mathbf{r}(t)=\langle t, t^2\rangle, 0\le t\le2.$$  
29. Compute arc length of helix $$\mathbf{r}(t)=\langle a\cos t, a\sin t, bt\rangle, 0\le t\le2\pi.$$  
30. Find unit binormal for $$\mathbf{r}(t)=\langle\cos t, \sin t, t\rangle.$$

## III. Gradient, Divergence, Curl (31–45)

31. Compute gradient of $$f(x,y,z)=x^2+y^2+z^2.$$  
32. Compute gradient of $$f(x,y,z)=xyz.$$  
33. Find divergence of $$\mathbf{F}=\langle x,y,z\rangle.$$  
34. Find divergence of $$\mathbf{F}=\langle yz,xz,xy\rangle.$$  
35. Compute curl of $$\mathbf{F}=\langle yz, xz, xy\rangle.$$  
36. Compute curl of $$\mathbf{F}=\langle y,-x,0\rangle.$$  
37. Compute divergence of $$\mathbf{F}=\langle e^x, e^y, e^z\rangle.$$  
38. Find points where $$\nabla f=0$$ for $$f=x^3+3y^2+z^2.$$  
39. Verify $$\nabla\cdot(\nabla\times\mathbf{F})=0.$$  
40. Verify $$\nabla\times(\nabla f)=0.$$  
41. Compute curl of $$\mathbf{F}=\langle y, -x, 0\rangle.$$  
42. Evaluate divergence of $$\mathbf{F}=\langle x^2, y^2, z^2\rangle.$$  
43. Compute gradient of $$f(x,y,z)=e^{x+y+z}.$$  
44. Compute curl of $$\mathbf{F}=\langle 2xy, 3yz, 4zx\rangle.$$  
45. Find divergence and curl of $$\mathbf{F}=\langle yz^2, xz^2, xy^2\rangle.$$

## IV. Line Integrals (46–60)

46. Evaluate $$\int_C (x+y)\,dx + (y-x)\,dy,$$ $$C:$$ line from (0,0) to (1,1).  
47. Evaluate $$\int_C x\,dy - y\,dx,$$ $$C:$$ circle $x^2+y^2=1.$$
48. Compute $$\int_C (y\,dx + x\,dy),$$ $$C:$$ path (0,0)→(1,0)→(1,1).  
49. Evaluate work done by $$\mathbf{F}=\langle y, x\rangle$$ along $$y=x^2, 0\le x\le1.$$  
50. Evaluate $$\int_C (x^2+y^2)\,ds$$ along line $$y=2x, 0\le x\le1.$$  
51. Compute $$\int_C \mathbf{F}\cdot d\mathbf{r}, \mathbf{F}=\langle x,y\rangle,$$ $$C: (0,0)\to(1,1).$$  
52. Evaluate $$\int_C y\,dx + x\,dy$$ along upper semicircle $$x^2+y^2=1.$$  
53. $$\mathbf{F}=\langle -y,x\rangle, C:$$ unit circle — find circulation.  
54. Evaluate work of $$\mathbf{F}=\langle xy, x+y\rangle$$ from (0,0) to (1,1).  
55. Find if $$\mathbf{F}=\langle2xy, x^2\rangle$$ is conservative.  
56. If conservative, find potential function for above.  
57. Compute $$\int_C \mathbf{F}\cdot d\mathbf{r}, \mathbf{F}=\nabla(3x^2y),$$ path (0,0)→(1,1).  
58. Evaluate $$\int_C e^{x}\,dx + e^{y}\,dy, C:(0,0)\to(1,1).$$  
59. Compute $$\int_C (x+y)\,ds,$$ $$C: (x=t,y=t^2,0\le t\le1).$$  
60. Evaluate $$\int_C y\,dx + x\,dy$$ for $$C:$$ square with corners (0,0),(1,0),(1,1),(0,1).

## V. Surface Integrals & Flux (61–80)

61. Compute surface area of plane $$z=3x+2y, 0\le x,y\le1.$$  
62. Compute $$\iint_S z\,dS$$ for surface $$z=x^2+y^2, z\le1.$$  
63. Evaluate $$\iint_S \mathbf{F}\cdot\mathbf{n}\,dS, \mathbf{F}=\langle x,y,1\rangle,$$ plane $$z=1-x-y.$$  
64. Find flux of $$\mathbf{F}=\langle x,y,z\rangle$$ through sphere $$x^2+y^2+z^2=1.$$  
65. Compute $$\iint_S (x+y)\,dS, S:(z=4-x^2-y^2, z\ge0).$$  
66. Evaluate flux of $$\mathbf{F}=\langle x,0,0\rangle$$ through plane $$x=2.$$  
67. Compute surface integral over paraboloid $$z=x^2+y^2, 0\le z\le1.$$  
68. Evaluate $$\iint_S (x^2+y^2)\,dS,$$ sphere $$x^2+y^2+z^2=9.$$  
69. Find flux of $$\mathbf{F}=\langle yz,xz,xy\rangle$$ through cube $$0\le x,y,z\le1.$$  
70. Compute outward flux through cylinder $$x^2+y^2=1, 0\le z\le2.$$  
71. Evaluate $$\iint_S xz\,dS, S:(x+y+z=3)$$ in first octant.  
72. Flux of $$\mathbf{F}=\langle 0,0,z\rangle$$ through $$z=4-x^2-y^2.$$  
73. Compute $$\iint_S \mathbf{F}\cdot\mathbf{n}\,dS, \mathbf{F}=\langle y,-x,0\rangle,$$ disk $$x^2+y^2\le1.$$  
74. Find flux of $$\mathbf{F}=\langle xy,xz,yz\rangle$$ through cube of side 2.  
75. Evaluate surface integral over hemisphere $$z=\sqrt{1-x^2-y^2}.$$  
76. Compute flux through plane $$2x+y+z=4.$$  
77. Evaluate $$\iint_S x\,dS$$ for surface $$z=xy, 0\le x,y\le1.$$  
78. Compute flux of $$\mathbf{F}=\langle x,y,1\rangle$$ through cone $$z=\sqrt{x^2+y^2}.$$  
79. Evaluate $$\iint_S z\,dS$$ over $$z=1-x^2-y^2, z\ge0.$$  
80. Flux through cylinder $$x^2+y^2=9, 0\le z\le4,$$ for $$\mathbf{F}=\langle x,y,z\rangle.$$

## VI. Green’s, Divergence, and Stokes’ Theorems (81–100)

81. Verify Green’s Theorem for $$\mathbf{F}=\langle -y,x\rangle$$ over unit disk.  
82. Verify Green’s Theorem for $$\mathbf{F}=\langle y^2, x^2\rangle$$ over square $$0\le x,y\le1.$$  
83. Evaluate $$\oint_C (2x - y)\,dx + (x + y)\,dy, C:$$ boundary of region $$y=x^2$$ to $$y=x.$$  
84. Apply Green’s Theorem to find area of circle $$x^2+y^2=4.$$  
85. Verify Divergence Theorem for $$\mathbf{F}=\langle x,y,z\rangle$$ on cube $$[0,1]^3.$$  
86. Verify Divergence Theorem for $$\mathbf{F}=\langle xy,yz,zx\rangle$$ on unit cube.  
87. Use Divergence Theorem to find flux of $$\mathbf{F}=\langle x,y,z\rangle$$ through sphere $$x^2+y^2+z^2=9.$$  
88. Compute flux of $$\mathbf{F}=\langle yz,xz,xy\rangle$$ via Divergence Theorem on cube $$[0,1]^3.$$  
89. Verify Stokes’ Theorem for $$\mathbf{F}=\langle -y,x,0\rangle$$ over disk $$x^2+y^2\le1.$$  
90. Verify Stokes’ Theorem for $$\mathbf{F}=\langle z,y,x\rangle$$ over plane $$x+y+z=1.$$  
91. Compute line integral of $$\mathbf{F}=\langle y,-x,z\rangle$$ using Stokes’ Theorem.  
92. Apply Divergence Theorem to compute volume integral of divergence of $$\mathbf{F}=\langle 2x,2y,2z\rangle$$ over sphere radius 1.  
93. Verify that both sides of Stokes’ Theorem agree for $$\mathbf{F}=\langle 0,0,xy\rangle.$$  
94. Use Green’s Theorem to find work around ellipse $$\frac{x^2}{a^2}+\frac{y^2}{b^2}=1.$$  
95. Use Divergence Theorem to compute outward flux through cylinder $$x^2+y^2=1, 0\le z\le2.$$  
96. Compute circulation of $$\mathbf{F}=\langle -y,x,0\rangle$$ along unit circle in (xy)-plane.  
97. Apply Stokes’ Theorem to $$\mathbf{F}=\langle y,z,x\rangle$$ on upper hemisphere.  
98. Verify that Divergence Theorem gives same result as direct flux integral for $$\mathbf{F}=\langle x^2,y^2,z^2\rangle.$$  
99. Compute circulation around square for $$\mathbf{F}=\langle y^2,x^2\rangle$$ using Green’s Theorem.  
100. Verify all three theorems (Green, Stokes, Divergence) for $$\mathbf{F}=\langle y,z,x\rangle$$ on appropriate regions.
