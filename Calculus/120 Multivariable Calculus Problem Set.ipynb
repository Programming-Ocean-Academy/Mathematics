## I. Functions of Several Variables & Limits (1–10)

4. Determine if $$f(x,y)=\frac{x^2-y^2}{x^2+y^2}$$ is continuous at $$(0,0)$$  
5. Find where $$f(x,y)=\frac{x^2y}{x^2+y^2}$$ is continuous  
6. Check differentiability of $$f(x,y)=x^2y+y^3$$ at $$(1,2)$$  
7. Sketch level curves of $$f(x,y)=x^2+y^2$$ for $$f=1,4,9$$  
8. Sketch contour plot of $$f(x,y)=x^2-y^2$$  
9. Find and plot level surfaces of $$f(x,y,z)=x^2+y^2+z^2$$
