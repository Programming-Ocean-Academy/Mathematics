# Phase 5 — Applications of Derivatives in Analysis

---

## 1️⃣ Mean Value Theorem (MVT)

### Statement

If \( f(x) \) is  

1. continuous on \([a,b]\), and  
2. differentiable on \((a,b)\),

then there exists some \( c \in (a,b) \) such that  
$$
\boxed{f'(c) = \frac{f(b) - f(a)}{b - a}}
$$

### Meaning

At some point, the **instantaneous slope** equals the **average slope** over the interval.

**Special Case:**  
If \( f(a) = f(b) \), then \( f'(c) = 0 \) — flat tangent (Rolle’s Theorem).

**Example**

$$
f(x) = x^2, \quad a = 1, \quad b = 3
$$
$$
f'(x) = 2x, \quad \frac{f(3) - f(1)}{3 - 1} = \frac{9 - 1}{2} = 4
$$
Thus \( f'(c) = 4 \Rightarrow 2c = 4 \Rightarrow c = 2 \).

✅ Slope at \( x = 2 \) matches average slope.

## 2️⃣ Increasing / Decreasing Tests

Use the sign of \( f'(x) \):

| Behavior | Condition |
|-----------|------------|
| Increasing | \( f'(x) > 0 \) |
| Decreasing | \( f'(x) < 0 \) |
| Stationary / Critical | \( f'(x) = 0 \) or undefined |

Plot sign changes to find local extrema.

---

## 3️⃣ Concavity & Inflection Review

- \( f''(x) > 0 \Rightarrow \) concave up  
- \( f''(x) < 0 \Rightarrow \) concave down  
- \( f''(x) \) changes sign → **inflection point**

**Example**

\( f(x) = x^3 - 3x^2 + 2 \)
$$
f'(x) = 3x^2 - 6x, \quad f''(x) = 6x - 6
$$
Setting \( f''(x) = 0 \Rightarrow x = 1 \).  
Concave down for \( x < 1 \), concave up for \( x > 1 \).  
Inflection point at \( (1, 0) \).

## 4️⃣ Curve Sketching Strategy

| Step | What to Find | Tool |
|------|---------------|------|
| 1 | Domain, intercepts | algebra |
| 2 | Vertical / horizontal asymptotes | limits |
| 3 | Increasing / decreasing | sign of \( f'(x) \) |
| 4 | Max / min | sign changes of \( f'(x) \) |
| 5 | Concavity / inflection | sign of \( f''(x) \) |
| 6 | End behavior | \( \lim_{x\to\pm\infty} f(x) \) |

This process gives a full qualitative picture of any function.

## 5️⃣ L’Hôpital’s Rule

### When to Use

If a limit produces an **indeterminate form**:
$$
\frac{0}{0} \quad \text{or} \quad \frac{\infty}{\infty}
$$
then:
$$
\boxed{\lim_{x \to a} \frac{f(x)}{g(x)} = \lim_{x \to a} \frac{f'(x)}{g'(x)}}
$$
(provided the new limit exists).

---

### Examples

**(a)**  
$$
\lim_{x \to 0} \frac{\sin x}{x} = \lim_{x \to 0} \frac{\cos x}{1} = 1
$$

**(b)**  
$$
\lim_{x \to \infty} \frac{3x^2 + 1}{5x^2 + 4x} = \lim_{x \to \infty} \frac{6x}{10x + 4} = \frac{6}{10} = \tfrac{3}{5}
$$

---

### Other Indeterminate Forms

| Type | How to Handle |
|-------|----------------|
| \( 0 \cdot \infty \) | Rewrite as fraction |
| \( \infty - \infty \) | Use common denominator |
| \( 0^0, 1^\infty, \infty^0 \) | Take logarithm, then use rule |

---

## 6️⃣ Derivative Tests Summary

| Test | Uses | Procedure |
|-------|------|------------|
| 1st Derivative | Increasing/decreasing + extrema | Check sign change of \( f' \) |
| 2nd Derivative | Concavity, confirm max/min | \( f'' > 0 \Rightarrow \) min; \( f'' < 0 \Rightarrow \) max |
| MVT | Existence of matching slope | \( f'(c) = \frac{f(b)-f(a)}{b-a} \) |
| L’Hôpital | Indeterminate limits | Replace \( f/g \) with \( f'/g' \) |

## 7️⃣ Example: Comprehensive Curve Analysis

Let \( f(x) = \dfrac{x^3 - 3x^2}{x - 1} \).

1. Domain: \( x \neq 1 \)  
2. Vertical asymptote: \( x = 1 \)  
3. No horizontal asymptote — degree difference = 1 → slant asymptote  
4. Derivative: find \( f'(x) \) via quotient rule  
5. Critical points: solve \( f'(x) = 0 \)  
6. Concavity: determine sign of \( f''(x) \)  
7. Combine to sketch a full behavior chart.

---

## 8️⃣ Conceptual Takeaways

- Derivatives **describe** how a function behaves.  
- \( f' \) shows **direction**; \( f'' \) shows **shape**.  
- MVT guarantees some \( c \) with slope equal to the average slope.  
- L’Hôpital’s Rule resolves **indeterminate limits**.
