## 1ï¸âƒ£ What Is a Differential Equation (DE)?

A **differential equation** relates a function to one or more of its **derivatives**.

$$
\boxed{\text{Differential Equation: } F(x, y, y', y'', \dots) = 0}
$$

It expresses *how a quantity changes* rather than *what the quantity is*.

---

### Examples

| Type | Example | Meaning |
|------|----------|----------|
| First-order | \( \frac{dy}{dx} = 3x^2 \) | slope of y depends on x |
| Second-order | \( y'' + y = 0 \) | describes oscillation |
| Growth/decay | \( y' = ky \) | exponential model |

## 3ï¸âƒ£ Separable Differential Equations

If the equation can be written as:
$$
\frac{dy}{dx} = g(x)h(y)
$$
then we can **separate variables**:
$$
\frac{1}{h(y)}\,dy = g(x)\,dx
$$
and integrate both sides.

---

### Example 1
$$
\frac{dy}{dx} = 2xy
$$
Separate:
$$
\frac{1}{y}\,dy = 2x\,dx
$$
Integrate:
$$
\ln|y| = x^2 + C
$$
Simplify:
$$
\boxed{y = Ce^{x^2}}
$$

## 4ï¸âƒ£ Exponential Growth & Decay

Form:
$$
\frac{dy}{dt} = ky
$$
Solution:
$$
\boxed{y = Ae^{kt}}
$$

* \( k > 0 \): growth  
* \( k < 0 \): decay

---

### Example 2

Population \( P \) grows at rate proportional to its size:
$$
\frac{dP}{dt} = 0.03P
\Rightarrow P = P_0 e^{0.03t}
$$
If \( P_0 = 100 \), after 10 years:
$$
P = 100 e^{0.3} \approx 135
$$

## 5ï¸âƒ£ Newtonâ€™s Law of Cooling

$$
\frac{dT}{dt} = -k(T - T_s)
$$
Solution:
$$
T(t) = T_s + (T_0 - T_s)e^{-kt}
$$
where \( T_s \) is surrounding temperature.

---

## 6ï¸âƒ£ Logistic Growth (Population with Limit)

$$
\frac{dP}{dt} = rP\left(1 - \frac{P}{K}\right)
$$
Solution:
$$
\boxed{P(t) = \frac{K}{1 + Ae^{-rt}}}
$$
where \( K \) is the carrying capacity.

## 9ï¸âƒ£ Solving First-Order Linear Equations

Form:
$$
\frac{dy}{dx} + P(x)y = Q(x)
$$

### Integrating Factor Method

Multiply both sides by:
$$
\mu(x) = e^{\int P(x)\,dx}
$$
Then:
$$
(\mu y)' = \mu Q(x)
$$
Integrate both sides:
$$
y = \frac{1}{\mu(x)} \left(\int \mu(x)Q(x)\,dx + C\right)
$$

---

### Example 3

$$
\frac{dy}{dx} + y = x
$$
Here \( P(x) = 1, \ Q(x) = x \):
$$
\mu = e^{\int 1\,dx} = e^x
$$
Then:
$$
(e^x y)' = xe^x
$$
Integrate RHS:
$$
e^x y = \int xe^x\,dx = e^x(x - 1) + C
$$
Simplify:
$$
\boxed{y = x - 1 + Ce^{-x}}
$$
